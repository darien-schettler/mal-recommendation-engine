```python
# Import the Necessary Libraries
from bs4 import BeautifulSoup
import urllib.request

import numpy as np
import pandas as pd

import time
```

```python
# specify the URL information to be concated (may be placeholding)

# Remains unchanged
page_root_1 = 'https://myanimelist.net/users.php?q='

# Will be defined as we go
search_word = ''

# Will remain unchanged
page_root_2 = '&loc=&agelow=0&agehigh=11000&g='

# Will have things added to it
page_constant = '&show='

# how many to index by
page_number = 24
```

```python
# Flag for first page
firstPage = True

# Flag for determining valid connection
connectionFlag = True

# Pick all the characters to be used in the triplet search
char_list = "abcdefghijklmnopqrstuvwxyz0123456789"
```

```python
# Initialize triplet list
char_triplets = []

# Make all the search triplets
for i in range(0,len(char_list)):
    for j in range(0,len(char_list)):
        for k in range(0, len(char_list)):
            char_triplets.append(char_list[i] + char_list[j] + char_list[k])
```

```python
mal_user_list = []
for triplet in char_triplets:

    connectionFlag = True
    firstPage = True
    search_word = str(triplet)

    while connectionFlag:

        quote_page_final = page_root_1 + search_word + page_root_2

        if not firstPage:
            quote_page_final += (page_constant+str(page_number))
            page_number += 24
        else:
            firstPage = False
            page_number = 24

        # See if the website still exists
        try:
            # query the website and return the html to the variable page
            page = urllib.request.urlopen(quote_page_final)
            print(quote_page_final)
        except:
             connectionFlag = False

        # parse the html using beautiful soup and store in variable `soup`
        soup = BeautifulSoup(page, 'html.parser')

        # Find the users on the page and put them into a list
        user_list = soup.findAll('div', attrs={'style' : 'margin-bottom: 7px;'})

        # Iterate through the user_list and append it to the overall user_list
        for user in user_list:
            mal_user_list.append(user.text)

    print("\n\nCompleted triplet {}".format(triplet))
    print("User List is Currently {} People Long".format(len(mal_user_list)))

    time.sleep(0.05)

    np.save('./data/user_info/mal_user_list.npy', mal_user_list)
```
